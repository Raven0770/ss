```python
# program 1

import numpy as np
import pandas as pd

# try:
data = pd.read_csv('house_prices.csv')
house_prices = data['Price']
# except FileNotFoundError:
#     print("Error: 'house_prices.csv' not found.")
#     print("Please make sure the CSV file is in the same directory as the script.")
#     exit()

q1 = np.percentile(house_prices, 25)
q3 = np.percentile(house_prices, 75)

iqr = q3 - q1

print(f"Sample House Prices:\n{house_prices.to_string(index=False)}")
print("-" * 30)
print(f"25th Percentile (Q1): ${q1:,.2f}")
print(f"75th Percentile (Q3): ${q3:,.2f}")
print(f"Interquartile Range (IQR): ${iqr:,.2f}")
```

```python
# Program 2
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load shopping trends dataset
shopping_df = pd.read_csv("shopping_trends.csv")

# Bar plot
plt.figure(figsize=(6,4))
sns.countplot(x="Customer Satisfaction", hue="Repeat Purchase", data=shopping_df)
plt.title("Customer Satisfaction vs Repeat Purchases")
plt.show()

# Stacked bar chart
cross_tab = pd.crosstab(shopping_df["Customer Satisfaction"], shopping_df["Repeat Purchase"])
cross_tab.plot(kind="bar", stacked=True, figsize=(6,4))
plt.title("Stacked Bar Chart: Satisfaction vs Repeat Purchases")
plt.ylabel("Number of Customers")
plt.show()
```

```python
# Program 3

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cars dataset
cars_df = pd.read_csv("CarsData.csv")

# Select only needed variables
subset = cars_df[["engineSize", "mpg", "price"]]

# Pair plot
sns.pairplot(subset)
plt.suptitle("Pair Plot of Car Attributes", y=1.02)
plt.show()
print("\n")

# Correlation heatmap
plt.figure(figsize=(6,4))
sns.heatmap(subset.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Car Attributes")
plt.show()
```

```python
# Program 4

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(42)

# Step 1: Create a skewed population distribution (e.g., log-normal to simulate skewed salaries)
population_size = 100_000
population_salaries = np.random.lognormal(mean=10, sigma=0.5, size=population_size)

# Step 2: Draw 10 random samples, each of size 50, and compute sample means
sample_means = []
sample_size = 50
num_samples = 10

for _ in range(num_samples):
    sample = np.random.choice(population_salaries, size=sample_size, replace=False)
    sample_means.append(np.mean(sample))

# Step 3: Plot histogram of sample means
plt.figure(figsize=(10, 6))
sns.histplot(sample_means, bins=8, kde=True, color='skyblue', edgecolor='black')
plt.title('Distribution of Sample Means (10 samples of size 50)', fontsize=14)
plt.xlabel('Sample Mean Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Program 5

import numpy as np
from scipy import stats

n = 20
sample_mean = 8
sample_std = 2
mu_0 = 0

t_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

df = n - 1

p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

alpha = 0.05
t_crit = stats.t.ppf(1 - alpha/2, df)

print(f"t-statistic = {t_stat:.2f}")
print(f"Degrees of freedom = {df}")
print(f"Critical value () = {t_crit:.3f}")
print(f"p-value = {p_value:.6f}")

if abs(t_stat) > t_crit:
    print("Reject H0: The drug significantly affects heart rate.")
else:
    print("Fail to reject H0: No significant effect detected.")
```
